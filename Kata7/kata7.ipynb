```python
#Declaracion de variables
new_planet =""
planets =[]

```

# Crea un ciclo while

```python
while new_planet != 'done':
    if new_planet:
        planets.append(new_planet)
    new_planet = input("Ingresa un nuevo planeta")

print (planets);
        
    
```
